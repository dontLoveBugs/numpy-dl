```python
import numpy as np
from utils import distance

# random to generate data
x = np.zeros((100, 2))

for i in range(100):
    if i & 1:
        x[i][0] = np.random.normal(2, scale=0.5)
        x[i][1] = np.random.normal(2, scale=0.5)
    else:
        x[i][0] = np.random.normal(0, scale=0.5)
        x[i][1] = np.random.normal(0, scale=0.5)

# visualization
import matplotlib.pyplot as plt  

plt.scatter(x[::, 0], x[::, 1])
plt.show()

# settings
k = 2
d = 2
```

```python
from kmeans import Kmeans
kmeans = Kmeans(k)
kmeans._randCentroids(x)

# visualization
plt.scatter(x[::, 0], x[::, 1])
plt.plot(kmeans.centroids[::, 0], kmeans.centroids[::, 1], "ro")
plt.show()
```

```python
# iterate to solve
kmeans.fit(x)

# visualization
plt.scatter(x[::, 0], x[::, 1])
plt.plot(kmeans.centroids[::, 0], kmeans.centroids[::, 1], "ro")
plt.show()
```

```python
test_data = np.array([[1, 1]])
labels = kmeans.predict(test_data)
print(labels, kmeans.centroids[labels])
```
